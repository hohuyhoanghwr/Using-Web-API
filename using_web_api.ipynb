```python
#Add in your Jupyter Notebook a new cell for loading the Alpha Vantage API Token:
import os
from dotenv import load_dotenv
import pandas as pd

load_dotenv()
av_token = os.getenv('AV_TOKEN')
```

```python
# Stock Ticker Symbol
symbol = 'AAPL'

url = f'https://www.alphavantage.co/query?function=INCOME_STATEMENT&symbol={symbol}&apikey={av_token}'
r = requests.get(url)
data = r.json()
data
```

```python
income_df = pd.DataFrame(data['annualReports'])
income_df
```

```python
income_df['fiscalDateEnding'] = pd.to_datetime(income_df['fiscalDateEnding'])
income_df['fiscalYear'] = income_df['fiscalDateEnding'].dt.year
cols_to_float = ['grossProfit', 'totalRevenue',
       'costOfRevenue', 'costofGoodsAndServicesSold', 'operatingIncome',
       'sellingGeneralAndAdministrative', 'researchAndDevelopment',
       'operatingExpenses', 'investmentIncomeNet', 'netInterestIncome',
       'interestIncome', 'interestExpense', 'nonInterestIncome',
       'otherNonOperatingIncome', 'depreciation',
       'depreciationAndAmortization', 'incomeBeforeTax', 'incomeTaxExpense',
       'interestAndDebtExpense', 'netIncomeFromContinuingOperations',
       'comprehensiveIncomeNetOfTax', 'ebit', 'ebitda', 'netIncome']

income_df[cols_to_float] = income_df[cols_to_float].apply(pd.to_numeric, errors='coerce')
income_df = income_df.sort_values(by='fiscalYear', ascending=True)
income_df
```

```python
# Plotting the total revenue of the company over the fiscal years
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(10, 6))
income_df.plot(
    x='fiscalYear',
    y='totalRevenue',
    kind='bar',
    ax=ax,
    title=f'Revenue of {symbol}',
    ylabel='Revenue (Billion USD)',
    xlabel='Fiscal Year',
    legend=False
)

# Format y-axis to billions
ax.yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{x/1e9:.0f}'))

plt.tight_layout()
plt.show()
```

```python
# Daily adjusted full time series of stock price 
symbol = 'AAPL'

url = f'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol={symbol}&apikey={av_token}'
r_price = requests.get(url)
data_price = r_price.json()
data_price
```

```python
# 2. Extract the time series data
time_series = data_price["Time Series (Daily)"]

# 3. Convert to DataFrame
df_price = pd.DataFrame.from_dict(time_series, orient="index")
df_price.index = pd.to_datetime(df_price.index)
df_price = df_price.sort_index()  # make sure it's in chronological order

# Optional: Rename columns for clarity
df_price = df_price.rename(columns={
    "1. open": "Open",
    "2. high": "High",
    "3. low": "Low",
    "4. close": "Close",
    "5. volume": "Volume"
})
df_price[["Open", "High", "Low", "Close","Volume"]] = df_price[["Open", "High", "Low", "Close","Volume"]].astype(float)

# 4. Plot
plt.figure(figsize=(10, 6))
plt.plot(df_price.index, df_price["Close"], label="Close Price")
plt.title(f"{symbol} Daily Close Price")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Plotting candlestick chart using mplfinance
import mplfinance as mpf
mpf.plot(df_price, type='candle', style='yahoo', volume=True, title=f"{symbol} Candlestick Chart")
```
